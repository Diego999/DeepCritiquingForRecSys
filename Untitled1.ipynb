```python
df_baseline = pd.read_csv('tables/CD_final/baselines/final_baselines.csv')
```

```python
df_baseline = df_baseline.drop(['rank','alpha','lambda','iter','similarity','corruption','root','topK'], axis=1)
```

```python
import ast
def remove_val(string):
    l = ast.literal_eval(string)
    return l[0]
```

```python
columns = list(df_baseline.columns)
```

```python
columns.pop(0)
```

```python
columns
```

```python
for c in columns:
    df_baseline[c] = df_baseline[c].apply(lambda x: remove_val(x))
```

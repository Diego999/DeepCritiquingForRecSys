# final explanation

```python
folder_path = 'tables/explanation/beer/final_explanation/'
metric='NDCG@10'
```

```python
csv_files = [join(folder_path, f) for f in listdir(folder_path) 
             if isfile(join(folder_path, f)) and f.endswith('.csv')]
```

```python
csv_files
```

```python
df_list = []

for filename in csv_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    df_list.append(df)

df = pd.concat(df_list, axis=0, ignore_index=True)
```

```python
df
```

```python
import ast

df['MAP@10'] = df['MAP@10'].apply(ast.literal_eval)
df['MAP@15'] = df['MAP@15'].apply(ast.literal_eval)
df['MAP@20'] = df['MAP@20'].apply(ast.literal_eval)
df['MAP@5'] = df['MAP@5'].apply(ast.literal_eval)
df['NDCG@10'] = df['NDCG@10'].apply(ast.literal_eval)
df['NDCG@15'] = df['NDCG@15'].apply(ast.literal_eval)
df['NDCG@20'] = df['NDCG@20'].apply(ast.literal_eval)
df['NDCG@5'] = df['NDCG@5'].apply(ast.literal_eval)
df['Precision@10'] = df['Precision@10'].apply(ast.literal_eval)
df['Precision@15'] = df['Precision@15'].apply(ast.literal_eval)
df['Precision@20'] = df['Precision@20'].apply(ast.literal_eval)
df['Precision@5'] = df['Precision@5'].apply(ast.literal_eval)
# df['Precision@10'] = df['Precision@10'].apply(ast.literal_eval)
# df['MAP@10'] = df['MAP@10'].apply(ast.literal_eval)
```

```python
df.columns
```

```python
df.groupby('model', as_index=False)['MAP@10', 'MAP@15', 'MAP@20', 'MAP@5', 
                                    'NDCG@10', 'NDCG@15', 'NDCG@20', 'NDCG@5', 
                                    'Precision@10', 'Precision@15', 'Precision@20', 'Precision@5',
                                    'Recall@10', 'Recall@15', 'Recall@20', 
                                    'Recall@5'].agg({'MAP@10': 'mean', 'MAP@15': 'mean', 
                                                     'MAP@20': 'mean', 'MAP@5': 'mean', 
                                                     'NDCG@10': 'mean',
                                                     'NDCG@15': 'mean', 'NDCG@20': 'mean', 
                                                     'NDCG@5': 'mean',
                                                     'Precision@10': 'mean', 'Precision@15': 'mean',
                                                     'Precision@20': 'mean', 'Precision@5': 'mean',
                                                     'Recall@10': 'mean', 'Recall@15': 'mean',
                                                     'Recall@20': 'mean','Recall@5': 'mean'})
```

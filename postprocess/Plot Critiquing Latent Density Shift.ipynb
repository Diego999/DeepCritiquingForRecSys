```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from utils.io import load_yaml
```

```python
df = pd.read_csv('tables/CD_Critiquing_Latent.csv')
```

```python
def show_density(df, name="CVNCF_DENSITY", x='X', y='Y', hue='stage', model='CVNCF', save=True):
    df = df[df['model']==model]
    fig, ax = plt.subplots(figsize=(4, 4))
    ax = sns.scatterplot(x='X', y='Y', hue='stage', style='stage', data=df)
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles=handles[1:], labels=labels[1:])
    #ax.legend(title="Legend Title")
    plt.tight_layout()
    if save:
        fig_path = load_yaml('config/global.yml', key='path')['figs']
        fig.savefig('{0}/{1}.pdf'.format(fig_path, name),
                    bbox_inches="tight", pad_inches=0, format='pdf')
    else:
        plt.show()
```

```python
show_density(df, name="CVNCF_DENSITY", x='X', y='Y', hue='stage', model='CVNCF', save=True)
```

```python
show_density(df, name="CNCF_DENSITY", x='X', y='Y', hue='stage', model='CNCF', save=True)
```

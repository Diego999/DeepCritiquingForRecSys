# Beer

```python
df = pd.read_csv('tables/beer_Critiquing_MAP.csv')
df
```

```python
df_cvncf = df[df['model'] == 'CE-VNCF']
df_cvncf
```

```python
df_cvncf_5 = pd.DataFrame.from_dict({'Metric': ['F-MAP@5']*80000, 
                        'Model': ['CE-VNCF']*80000, 
                        'F-MAP': np.random.normal(df_cvncf.loc[0, 'F-MAP@5'], df_cvncf.loc[1, 'F-MAP@5']/1.96, 80000)})

df_cvncf_10 = pd.DataFrame.from_dict({'Metric': ['F-MAP@10']*80000, 
                        'Model': ['CE-VNCF']*80000, 
                        'F-MAP': np.random.normal(df_cvncf.loc[0, 'F-MAP@10'], df_cvncf.loc[1, 'F-MAP@10']/1.96, 80000)})

df_cvncf_20 = pd.DataFrame.from_dict({'Metric': ['F-MAP@20']*80000, 
                        'Model': ['CE-VNCF']*80000, 
                        'F-MAP': np.random.normal(df_cvncf.loc[0, 'F-MAP@20'], df_cvncf.loc[1, 'F-MAP@20']/1.96, 80000)})
```

```python
df_cncf = df[df['model'] == 'CE-NCF']
df_cncf
```

```python
df_cncf_5 = pd.DataFrame.from_dict({'Metric': ['F-MAP@5']*80000, 
                        'Model': ['CE-NCF']*80000, 
                        'F-MAP': np.random.normal(df_cncf.loc[2, 'F-MAP@5'], df_cncf.loc[3, 'F-MAP@5']/1.96, 80000)})

df_cncf_10 = pd.DataFrame.from_dict({'Metric': ['F-MAP@10']*80000, 
                        'Model': ['CE-NCF']*80000, 
                        'F-MAP': np.random.normal(df_cncf.loc[2, 'F-MAP@10'], df_cncf.loc[3, 'F-MAP@10']/1.96, 80000)})

df_cncf_20 = pd.DataFrame.from_dict({'Metric': ['F-MAP@20']*80000, 
                        'Model': ['CE-NCF']*80000, 
                        'F-MAP': np.random.normal(df_cncf.loc[2, 'F-MAP@20'], df_cncf.loc[3, 'F-MAP@20']/1.96, 80000)})
```

```python
makeup_df = df_cvncf_5.append(df_cvncf_10).append(df_cvncf_20).append(df_cncf_5).append(df_cncf_10).append(df_cncf_20)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
makeup_df
```

```python
makeup_df['Metric'] = makeup_df['Metric'].apply(lambda m: m.replace('F-MAP@', 'K='))
```

```python
# fig9 = plt.figure(figsize=(3, 3), dpi=100)
# ax9 = fig9.add_subplot(111)
fig, ax = plt.subplots(figsize=(3, 3), dpi=100)
ax = sns.barplot(x="Metric", y='F-MAP', hue='Model', data=makeup_df, palette="Set3", edgecolor=".2")
legend = plt.legend(loc='lower left', shadow=True, fontsize=10)
ax.set_xlabel("Number of Top-K Key Phrases")
# legend = plt.legend(shadow=True, fontsize=10)
handles, labels = ax.get_legend_handles_labels()
plt.tight_layout()
fig_path = 'figs'
name = 'falling_map_beer'
fig.savefig('{0}/{1}.pdf'.format(fig_path, name), bbox_inches="tight", pad_inches=0, format='pdf')
```

```python
fig, ax = plt.subplots(figsize=(4, 4))
ax = sns.boxplot(x="Metric", y='Value', hue='Model', data=makeup_df, palette="Set3", width=0.7)
```

```python
fig, ax = plt.subplots(figsize=(4, 4))
ax = sns.violinplot(x="Metric", y='Value', hue='Model', data=makeup_df, palette="Set3")
```

```python
df = pd.read_csv('tables/CD_Critiquing_MAP.csv')
df
```

```python
df_map5 = df[['model', 'F-MAP@5']]
df_map5.loc[:, 'Metric'] = ['F-MAP@5']*len(df_map5)
df_map5.rename(columns={'F-MAP@5':'F-MAP', 'model':'Model'}, inplace=True)
df_map5
```

```python
df_map10 = df[['model', 'F-MAP@10']]
df_map10.loc[:, 'Metric'] = ['F-MAP@10']*len(df_map10)
df_map10.rename(columns={'F-MAP@10':'F-MAP', 'model':'Model'}, inplace=True)
df_map10
```

```python
df_map20 = df[['model', 'F-MAP@20']]
df_map20.loc[:, 'Metric'] = ['F-MAP@20']*len(df_map20)
df_map20.rename(columns={'F-MAP@20':'F-MAP', 'model':'Model'}, inplace=True)
df_map20
```

```python
df_map5 = df_map5.sort_values('Model', ascending=False)
df_map10 = df_map10.sort_values('Model', ascending=False)
df_map20 = df_map20.sort_values('Model', ascending=False)
```
